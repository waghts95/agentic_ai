```python
from langchain_ollama.chat_models import ChatOllama
from langchain.agents import AgentExecutor, create_tool_calling_agent
#from langchain_mcp.adapters.ollama import OllamaMcpClient
from langchain_core.prompts import ChatPromptTemplate
from langchain_mcp.client import McpClient
```

```python
# 1. Configuration (Ollama and MCP Server URL)
OLLAMA_MODEL = "llama3.1"
OLLAMA_URL = "http://localhost:11434"
# *** CHANGE 2: Define the fixed URL for the running server ***
MCP_SERVER_URL = "http://localhost:8080" 
```

```python
# 2. Define the MCP Server connection information
# *** CHANGE 3: We provide the URL, not a launch command ***
server_connection = {
    "transport": "http",
    "url": MCP_SERVER_URL
}
```

```python

# 3. Initialize Ollama LLM and MCP Client
llm = ChatOllama(model=OLLAMA_MODEL, base_url=OLLAMA_URL)

# Use the connection dictionary to initialize the client
mcp_client = OllamaMcpClient.from_server_config(server_connection)
```

```python

# 4. Agent setup (same as before)
tools = mcp_client.get_tools()

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an expert math assistant. You must use the available tools for calculations."),
    ("human", "{input}")
])

agent = create_tool_calling_agent(llm, tools, prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools)
```

```python

# 5. Run the query
query = "What is the sum of 123 and 456? Provide the final answer."
print(f"User: {query}")

# The agent decides to call the 'add_numbers' tool over HTTP
response = agent_executor.invoke({"input": query})

print(f"\nAI Response:\n{response['output']}")
```
