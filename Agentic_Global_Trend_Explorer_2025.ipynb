## ðŸ§© Objective

Conduct a structured, data-driven analysis of <b>global AI investment trends</b> in 2025.<br>
Identify <b>key countries or regions</b> leading AI investment, quantify <b>growth since 2023</b><br> 
Provide <b>insightful reflections</b> on global investment patterns.

## ðŸ§  Task Breakdown (Agent Steps)

### **1. Information Gathering**
- Search for the most recent (2025) reports or news about **AI investments** in major economies:
  - United States  
  - Europe (EU as a bloc)  
  - China  
  - India  
  - Japan  
  - Other emerging regions (e.g., Southeast Asia, Middle East)

---

### **2. Summarization**
- For each region, produce a concise summary (2â€“3 sentences) covering:
  - Current AI investment amount or trend (2025)
  - Primary drivers (e.g., private funding, government policy, startups, defense, etc.)
  - Any notable sector focus (e.g., generative AI, robotics, healthcare AI)

---

### **3. Quantitative Estimation**
Using **hypothetical 2023â€“2025 investment data**, calculate the **percentage growth** for each region:

| Region | 2023 Investment (USD B) | 2025 Investment (USD B) |
|--------|--------------------------|--------------------------|
| United States | 120 | 165 |
| Europe | 80 | 100 |
| China | 150 | 210 |
| India | 25 | 45 |
| Japan | 30 | 40 |

*(You can extend this table to include additional regions if the agent finds relevant data.)*

---

### **4. Comparative Analysis**
- Rank all regions by **growth rate (%)**.  
- Identify:
  - Fastest-growing AI investment markets  
  - Mature markets with stable or saturated growth  
  - Emerging economies showing acceleration  

---

### **5. Reflection and Insights**
- Provide a **global synthesis paragraph** discussing:
  - The geographic shift of AI funding (e.g., Asia vs West)  
  - The balance between private vs public investment  
  - The overall direction of global AI innovation capital in 2025  

```python
# Description: Agentic AI example performing multi-step reasoning, data lookup, and reflection.
# --------------------------------------------------------
# 1. Import dependencies
# --------------------------------------------------------
from langchain_ollama import OllamaLLM
from langchain.agents import initialize_agent, AgentType
from langchain.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun
from langchain.memory import ConversationBufferMemory
import math, re
```

```python
# --------------------------------------------------------
# 2. Connect to local LLM
# --------------------------------------------------------
# Make sure Ollama is running: `ollama serve`
# And pull the model first: `ollama pull mistral`
llm = OllamaLLM(model="mistral")
```

```python
# --------------------------------------------------------
# 3. Define Tools
# --------------------------------------------------------

@tool
def summarize_text(text: str) -> str:
    """Summarizes long text into concise, readable key points."""
    sentences = re.split(r'(?<=[.!?]) +', text)
    summary = " ".join(sentences[:5])
    return f"Summary: {summary}"

@tool
def calculate_growth(input_text: str) -> str:
    """
    Calculates percentage growth based on input like 'from 120 to 165'.
    Example: 'from 120 to 165' â†’ 37.5% growth.
    """
    try:
        numbers = [float(x) for x in re.findall(r"\d+\.?\d*", input_text)]
        if len(numbers) < 2:
            return "Invalid input. Use format: 'from X to Y'"
        start, end = numbers[0], numbers[1]
        growth = ((end - start) / start) * 100
        return f"{growth:.2f}% growth"
    except Exception as e:
        return f"Error calculating growth: {e}"

@tool
def reflect_global_trends(text: str) -> str:
    """
    Reflect on global AI investment distribution and balance.
    Checks if growth patterns indicate regional convergence or divergence.
    """
    prompt = f"""
    Analyze this AI investment summary:
    {text}

    Reflect concisely:
    - Which regions are accelerating fastest?
    - Are investments globally balanced or uneven?
    - What might this imply for global AI leadership in 2025?
    """
    reflection = llm.invoke(prompt)
    return reflection

# --------------------------------------------------------
# 4. Search Tool & Memory
# --------------------------------------------------------
web_search = DuckDuckGoSearchRun()
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)
```

```python
# --------------------------------------------------------
# 5. Combine Tools into Agent
# --------------------------------------------------------
tools = [summarize_text, calculate_growth, reflect_global_trends, web_search]


agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent_type=AgentType.CHAT_ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
    handle_parsing_errors="auto",
    memory=memory
)
```

```python
# --------------------------------------------------------
# 6. Main Query â€” Global AI Investment (2025)
# --------------------------------------------------------
query = """
Perform a structured analysis of global AI investment in 2025.

Step 1: Search for the latest 2025 news or reports about AI investments in
the United States, Europe, China, India, and Japan.

Step 2: Summarize the investment trend in 2â€“3 lines per region.

Step 3: Using the following data, calculate the investment growth rate:
- US: from 120 to 165
- Europe: from 80 to 100
- China: from 150 to 210
- India: from 25 to 45
- Japan: from 30 to 40

Step 4: Combine the findings to describe which regions are growing fastest
and which are mature or slowing.

Step 5: Reflect globally on whether AI investment distribution is balanced,
shifting eastward, or dominated by certain economies.
"""
```

```python
# --------------------------------------------------------
# 7. Run the Agent
# --------------------------------------------------------
result = agent.run(query)

# --------------------------------------------------------
# 8. Display Final Output
# --------------------------------------------------------
print("\nðŸ§  Final Global AI Investment Analysis:\n")
print(result)
```
